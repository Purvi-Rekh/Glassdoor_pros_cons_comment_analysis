```python
import glob

csvfiles = []
for file in glob.glob("data-jobs-listings-glassdoor//*.csv"):
    csvfiles.append(file)
csvfiles
```

```python
for file in csvfiles:
    df=pd.read_csv(file)
    print("*************",file,"*********")
    print(df.head(10))
    
    
```

```python
def get_file_name():
    for file in csvfiles:
        yield file
    
```

```python
f=get_file_name()

def get_next_dataFrame():
    try:
        file=next(f)
        print("\n",file,"\n")
        df=pd.read_csv(file)
        print(df.head(10))
    except:
        print("last element")
```
